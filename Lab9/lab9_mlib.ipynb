```python
import networkx as nx
import matplotlib.pyplot as plt

# Sample boolean vectors (replace these with your actual data)
vectors = [
    [0, 1, 0],
    [1, 0,1]
]

# Calculate Hamming distance matrix
hamming_matrix = [[sum(a != b for a, b in zip(v1, v2)) for v2 in vectors] for v1 in vectors]

# Define a threshold
threshold = 2

# Create a graph
G = nx.Graph()

# Add nodes and edges based on the threshold
for i in range(len(vectors)):
    for j in range(i + 1, len(vectors)):
        if hamming_matrix[i][j] <= threshold:
                       G.add_edge(str(vectors[i]), str(vectors[j]))  # Convert tuples to strings


# Draw the circular graph
pos = nx.circular_layout(G)

nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8, edge_color='gray', font_color='black')

plt.show()
```
