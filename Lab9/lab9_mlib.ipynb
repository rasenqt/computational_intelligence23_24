```python
import networkx as nx
import matplotlib.pyplot as plt

# Sample boolean vectors (replace these with your actual data)
vectors = [
    [False, True, False],
    [True, False, True]
]

# Calculate Hamming distance matrix
hamming_matrix = [[sum(a != b for a, b in zip(v1, v2)) for v2 in vectors] for v1 in vectors]

# Define a threshold
threshold = 2

# Create a graph
G = nx.Graph()

# Add nodes and edges based on the threshold
for i in range(len(vectors)):
    for j in range(i + 1, len(vectors)):
        if hamming_matrix[i][j] <= threshold:
            G.add_edge(tuple(vectors[i]), tuple(vectors[j]))  # Use tuples directly

# Draw the graph with a spring layout and adjusted label size
plt.figure(figsize=(8, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_size=12, node_size=700, node_color='skyblue', font_color='black', font_weight='bold', edge_color='gray')

plt.show()
```
